Images should be of same size before feeding it to neural networks

If images are of different size then neural network gets confused

Therefore resizing the image is important

Resizing the image using pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/Hunter-350-AU-Tom-004.jpg')
img_resized = img.resize((200,200))
```

```python
#save the resized image
img_resized.save('dog_image_resized.jpg')
```

```python
# displaying the resized image from the numpy array

img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plt = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Convert RGB image to gray scale using OpenCV lbrary

```python
#import openCV library
import cv2
```

```python
img = cv2.imread('/content/Hunter-350-AU-Tom-004.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
greyscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(greyscale_image)
```

```python
greyscale_image.shape
```
